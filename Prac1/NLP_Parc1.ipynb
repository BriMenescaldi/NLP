```python
!pip install beautifulsoup4

import requests
from bs4 import BeautifulSoup
import re
```

```python
url='https://tableros.yvera.tur.ar/'
response = requests.get(url,verify=False)

#objeto BeautifulSoup con el contenido de la página
soup = BeautifulSoup(response.content, 'html.parser')

links = soup.find_all('a')

# buscar "tablero ODS"
link_tab_ODS = None
for link in links:
    if 'href' in link.attrs and 'tablero_ODS' in link['href']:
        link_tab_ODS = link['href']
        break


if link_tab_ODS:
    print("Found the link:", link_tab_ODS)

    pag_tab_ODS = requests.get(link_tab_ODS, verify=False)

    soup = BeautifulSoup(pag_tab_ODS.content, 'html.parser')

    parrafos = soup.find_all('p')

    #parrafos con texto y strip() sin espacios ni contenedores html
    for paragraph in parrafos:
       print(paragraph.text.strip())
```

```python
pdf_file_path = "/content/PROYECTO DE LEY TURISMO SOCIAL 2004.pdf"
with open(pdf_file_path, "rb") as pdf_file:
    #para lectura
    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # recorrido de las páginas
    for page_num in range(len(pdf_reader.pages)):

        page = pdf_reader.pages[page_num]

        #extracción texto
        text = page.extract_text()

        print("Pág", page_num + 1, ":", text.strip())
```
